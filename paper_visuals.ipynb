```python
import numpy as np
from matplotlib.colors import ListedColormap

from loss_vis import plot_selection_area
from afpo import AgeFitnessPareto
from util import create_diamond, create_hollow_circle, create_square
```

# Target shapes visual

```python
def get_target_shape(shape, resolution):
    """
    Returns the target shape using self.shape and the resolution of self.base_layer.
    Always returns 64x64 numpy array. 
    """

    target_size = 64 // resolution

    assert resolution in [2**n for n in range(int(np.log2(64)) + 1)]

    if shape == 'square':
        target = create_square(target_size)
    elif shape == 'diamond':
        target = create_diamond(target_size)
    elif shape == 'circle':
        target = create_hollow_circle(target_size)

    # Upsize back to 64x64 because that's how we're comparing 
    # while target.shape[0] < 64: 
    #     target = np.repeat(np.repeat(target, 2, axis=0), 2, axis=1)
    # assert target.shape[0] == 64

    return target

shape = 'circle'
data_64x64 = get_target_shape(shape, 1)
data_32x32 = get_target_shape(shape, 2)
data_16x16 = get_target_shape(shape, 4)
data_8x8 = get_target_shape(shape, 8)

dark_color = (0,0,0.7)
light_color = (1,1,1)

red = (1,0,0)
green = (0,0.5,0)
blue = (0,0,1)
orange = (1,0.5,0)
white = (1,1,1)

color_maps = [
    ListedColormap([blue, white]),
    ListedColormap([orange, white]),
    ListedColormap([green, white]),
    ListedColormap([red, white]),
]
plot_selection_area(data_64x64, data_32x32, data_16x16, data_8x8, color_maps)
```
